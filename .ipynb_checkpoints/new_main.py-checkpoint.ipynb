```python
def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = sp.interpolate.interp1d(logx, logy, kind=kind)
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp
    
def get_day(x):
    return x.day

def get_month(x):
    return x.month



def get_valo(P,K,Delta,N):
    #P = np.array(stripper.df['DF'])
    #K = np.array(stripper.df['K'])
    #Delta = np.array(stripper.df['Base'])
    #N = 1 #Nominal
    n = len(P)
    V = np.array([N *(P[n-1] +  np.sum((P*K*Delta)[i:] )) for i in range(n)])
    
    return V

def get_CF(P,K,Delta,N):
    n = len(P)
    CF = np.array([(N * P*K*Delta)[i] for i in range(n)])
    
    return CF
    

#TODO : Delta[i,j] matrix


@jit(nopython=True)
def polynome(p, P, K, Delta,V_N, i,LGD=.6):
    n = len(P)
    zero_target = P[n-1]*p**Delta[n-1] 
    zero_target += np.sum((Delta*P *((p**Delta[i]*K*Delta[i])+(p**(Delta[i]*(i))-p**(Delta[i]))*(1-LGD)))[i:]) 
    zero_target += - V_N[i]
    return zero_target

@jit(nopython=True)
def solve_polynome(P, K, Delta, V_N, i,epsilon = 1e-8):

    left = 0
    right= 1

    while right-left>epsilon :
        
        mid = (left+right) / 2
        if (polynome(left,P,K,Delta,V_N,i))*(polynome(mid,P,K,Delta,V_N,i)) < 0:
            right = mid
        else:
            left = mid
    
    return left


@jit(nopython=True)
def get_PS(P,K,Delta,V_N):
    return [solve_polynome(P,K,Delta,V_N,i) for i in range(len(P))]

class Stripper():
    def __init__(self,start='2021-01-06',end='2071-01-05',nominal = 100000, freq = 3): #freq = nb de mois
        self.data = pd.read_excel('Bloom_EUR_OIS.xlsx')
        self.start = int(pd.Timestamp(start).value/(24*3600*1000000000))
        self.end = int(pd.Timestamp(end).value/(24*3600*1000000000))
        self.days = np.linspace(self.start, self.end, self.end-self.start+1)
        
        
        self.tt = [pd.Timestamp(date).value/(24*3600*1000000000) for date in self.data['Payment Date']]

        self.DF   = np.vectorize(log_interp1d(self.tt, self.data['Discount']))(self.days)
        #self.TAUX = np.vectorize(log_interp1d(self.tt, -self.data['Market Rate']))(self.days)
        
        self.T    = np.vectorize(pd.Timestamp)(self.days*24*3600*1000000000)
        self.TAUX = [1/100 for _ in range(len(self.T))]
        self.N    = [nominal  for _ in range(len(self.T))]
        self.Comp = [freq*30.5/365 for _ in range(len(self.T))]
        
        self.df = pd.DataFrame({'date':self.T, 'DF':self.DF, 'K':self.TAUX,'N':self.N,'Base':self.Comp})
        
        self.df['days']   = self.df['date'].apply(get_day)
        self.df['months'] = self.df['date'].apply(get_month)
        
        self.df = self.df[self.df.days == 1]
        self.df = self.df[self.df.months%freq == 0]
        
        
         
        valo = get_valo(P = np.array(self.df['DF']),
                                K = np.array(self.df['K']),
                                Delta = np.array(self.df['Base']),
                                N = nominal)

        
        self.df['V'] = valo
        
        self.df['V/N'] = self.df['V']/nominal
        
        self.df['CF'] = get_CF(P = np.array(self.df['DF']),
                                       K = np.array(self.df['K']),
                                       Delta = np.array(self.df['Base']),N=nominal)
         
        self.df['PD'] = 1 -  np.array(get_PS(P = np.array(self.df['DF']),
                                       K = np.array(self.df['K']),
                                       Delta = np.array(self.df['Base']),
                                       V_N = np.array(self.df['V/N'])))
        
        
```
